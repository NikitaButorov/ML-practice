```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
axis.plot(x,x**2, label = 'X Squared',color = "yellow",marker="o",
         markerfacecolor="blue")
axis.plot(x,x**3,label = "X Cubed",color = "purple",lw=3,ls = ':',marker='o')
axis.set_xlim([0,3])
axis.set_ylim([0,10])
plt.legend()
plt.show()
```
