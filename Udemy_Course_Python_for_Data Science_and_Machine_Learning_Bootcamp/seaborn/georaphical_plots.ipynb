```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import pandas as pd
init_notebook_mode(connected=True)
```

```python
data = dict(type = 'choropleth',
           locations = ['AZ','CA','NY'],
           locationmode = 'USA-states',
           colorscale = 'Greens',
           text = ['text 1','text 2', 'text 3'],
           z= [1.0,2.0,3.0],
           colorbar = {'title': 'Colorbar Titel Goes Here'})
```

```python
data
```

```python
layout = dict(geo={'scope':'usa'})
```

```python
choromap = go.Figure(data = [data],layout=layout)
iplot(choromap)
```

```python
df = pd.read_csv('2011_US_AGRI_Exports')
df.head()
```

```python
exports_data = dict(type='choropleth',
                   colorscale='Jet',
                   locations = df['code'],
                   locationmode = 'USA-states',
                   z=df['total exports'],
                   text = df['text'],
                    marker = dict(line = dict(color = 'rgb(255,255,255)',width=2)),
                   colorbar={'title': 'Millions USD'})
exports_layout = dict(title = '2011_US_AGRI_Exports',geo = dict(scope='usa',showlakes = True,lakecolor='rgb(85,173,240)'))
```

```python
choromap2 = go.Figure(data = [exports_data],layout=exports_layout)
iplot(choromap2)
```

```python
world_df = pd.read_csv('2014_World_GDP')
```

```python
world_df.head()
```

```python
world_data = dict(type='choropleth',
                   locations = world_df['CODE'],
                   z = world_df['GDP (BILLIONS)'],
                   text = world_df['COUNTRY'],
                   colorbar={'title': 'GDP in Billions USD'})
world_layout = dict(title = '2014 GLOBAL GDP',
                   geo = dict(showframe = True,
                             projection = {'type':'kavrayskiy7'}))
```

```python
choromap3 = go.Figure(world_data,world_layout)
iplot(choromap3)
```
