```python
import seaborn as sns
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
tips.pivot_table(index='sex',columns="smoker")
```

```python
tc = tips.corr(numeric_only=True)
tc
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp = flights.pivot_table(index="month",values="passengers",columns="year")
```

```python
sns.heatmap(fp,cmap="magma",linecolor="white",lw=0.5)
```

```python
sns.clustermap(fp)
```
